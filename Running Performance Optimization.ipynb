```python
def find_exist(element, L):
    for e in L:
        if e == element:
            return e
    return False
```

## deque, double ending queue

```python
total_length = 1000
message_queue = [None] * total_length
```

```python
from collections import deque
```

```python
deque_msg = deque(message_queue)
```

```python
def msg_control(running=1000):
    for i in range(running):
        new_msg_num = random.randint(0, 10)
        while new_msg_num:
            #message_queue.pop(0)
            deque_msg.popleft()
            t = time.time()
            #message_queue.append(t)
            deque_msg.append(t)
            new_msg_num -= 1
```

```python
%timeit msg_control()
```

```python
@dataclass
class DCustomer: 
    name : str
    age : int
    rank : int
```

```python
from string import ascii_letters
```

```python
def first_upper(string):
    return string[0].upper() + string[1:].lower()
```

```python
def gene_name(): 
    length = random.randint(3, 6)
    return first_upper(''.join(random.sample(ascii_letters, length)))
```

## 近似化（往往伴随了随机化）

有很多复杂问题，如果要获得其精确解，那么需要的求解是非常复杂的。 如果我们要获得近似解，则，难度会降低很多~

## 数值分析(numerical analysis) 是进行算法优化非常重要的方式

### 2018有一个银行，要做运钞车的线路优化问题

```python
def fac(n):
    return n * fac(n - 1) if n > 1 else 1
```
