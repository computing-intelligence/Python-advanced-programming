```python
def pop_all(elements):
    # version-01 for merge functions
    while elements:
        if isinstance(elements, list):
            elements.pop(0)
        elif isinstance(elements, collections.deque):
            elements.popleft(0)
            
def pop_all_merged(elements, attr, arg=None): # version-02 for merge functions
    while elements:
        method = elements.__getattribute__(attr)
        if arg: method(arg)
        else: method()
```

```python
some_numbers = [random.randint(-100, 100) for _ in range(100000)]
```

```python
target_user_information = [191, 121, 170, 221, -19, 1198]
```

```python
def gen_random_user():
    return [random.randint(-200, 200) for _ in range(6)]
```

```python
other_usres = [
    gen_random_user() for _ in range(1000000)
]
```

## Our task is find the top-20 similar customers 

```python
def distance(x, y): 
    return np.sum((np.array(x) - np.array(y)) ** 2)
```

$$ distance =  \sum_{i \in D} (x_i - y_i) ^ 2 $$

```python
%time most_similar = sorted(other_usres, key=lambda o: distance(o, target_user_information))[:20]
```

```python
import heapq
```

```python
h_users = []
```

```python
other_usres = [
    gen_random_user() for _ in range(1000000)
]
```

```python
%time 

def push_all_user():
    for user in other_usres:
        heapq.heappush(h_users, (distance(user, target_user_information), user))
```

```python
%time push_all_user()
```

```python
%time most_similar=heapq.nsmallest(20, h_users)
```

```python
heapq.heapify(other_usres)
```

## OrderDict

Dict有一个特点：当我遍历dict的时候，输出的顺序和我们加入的顺序是不一定是一样的。

```python
agents = [3, 2, 1, 5, 4] # 顺序是打乱的

for a in agents:
    actions[a] =  'some'
# 对agent的行为做估计
```

```python
for agent in actions:
    a
    agent[a]
# 使用估计的action做计算， 做更新
# 更新的agent的顺序就不是
```

```python
from collections import OrderedDict
```

```python
help(OrderedDict)
```

```python
agents = [3, 1, 4, 5, 2]
```

```python
action_value = OrderedDict()
```

```python
for a in agents:
    action_value[a] = random.randint(-10, 10)
```

```python
action_value
```

```python
actions = [33, 12, 31, 12, 54]
```

```python
for a, v in action_value.items():
    if v > 0: actions[a-1] += v / 10
    else:
        actions[a-1] -= v/10
```
