## Assuming:  有N个class，需要添加N个 __repr__

````python
def add_repr_to_cls(cls):
    repr_string = f'{cls.__name__}'
    for attr in dir(cls):
        if callable(cls.__getattribute__(cls, attr)):
            pass
        elif not attr.startswith('__'):
            repr_string += f'\n{attr}:{cls.__getattribute__(cls, attr)}'
    
    cls.__repr__ = lambda s:  repr_string
    
    return cls
````

````python
def add_repr_to_instance(cls): 
    def get_attribute(obj):
        return {
            name: getattr(obj, name) for name in dir(obj)
            if not callable(getattr(obj, name)) and not name.startswith('__')
        }

    def __repr__(obj):
        cls_descrip = get_attribute(cls)
        cls_descrip.update(get_attribute(obj))
        
        description = '\n\t'.join([f'{key}: {value}' for key, value in cls_descrip.items()])
        
        infomation = f"""{cls.__name__}: 
        {description}
        """

        return infomation
    
    cls.__repr__ = __repr__ 
    
    return cls
````

````python
def is_camed(string): return string == inflection.camelize(string)
````

````python
inflection.underscore('SomeOtherName')
````

````python
def change_cammal_to_underscore(name, parent, attrs):
    new_attr = {
        inflection.underscore(a) : value
        for a, value in attrs.items() if is_camed(a) and callable(value)
    }
    
    attrs.update(new_attr)

    return type(name, parent, attrs)
````

````python
import math


class Planet:
    Number = 0

    def __init__(self, color, mass, position, v_init):
        Planet.Number += 1
        self.color = color
        self.mass = mass
        self.x, self. y = position
        self.vx, self.vy = v_init

    def gravity(self, other_planet):
        distance = math.sqrt((self.x - other_planet.x) ** 2 + (self.y - other_planet.y) ** 2)
        f = self.mass * other_planet.mass / (distance ** 2)
        fx = (other_planet.x - self.x) / distance * f
        fy = (other_planet.y - self.y) / distance * f
        return fx, fy

    def v_change(self, f, t=1e-3):
        fx, fy = f
        ax, ay = fx / self.mass, fy / self.mass
        self.vx = self.vx + ax * t
        self.vy = self.vy + ay * t

    def p_change(self, t=1e-3):
        self.x += self.vx * t
        self.y += self.vy * t
````
